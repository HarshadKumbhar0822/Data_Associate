```python
# Clustering
```

```python
# KMeans
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=3)
```

```python
kmeans.fit(iris_df)
```

```python
kmeans.labels_
```

```python
kmeans.cluster_centers_
```

```python
kmeans.predict(iris_df)
```

```python
plt.scatter(iris_df['sepal length (cm)'], iris_df['sepal width (cm)'], c=kmeans.labels_)
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], color='red')
plt.show()
```

```python
plt.scatter(iris_df['petal length (cm)'], iris_df['petal width (cm)'], c=kmeans.labels_)
plt.scatter(kmeans.cluster_centers_[:, 2], kmeans.cluster_centers_[:, 3], color='red')
plt.show()
```

```python
# Classification
from sklearn.linear_model import LogisticRegression
```

```python
iris_df
```

```python
iris_df_new = pd.DataFrame(data=data.data, columns=data.feature_names)
```

```python
iris_df_new
```

```python
iris_df_new['target'] = data.target
```

```python
iris_df_new
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(iris_df_new.drop('target', axis=1), iris_df_new['target'], test_size=0.2)
```

```python
print(len(X_train))
```

```python
lr = LogisticRegression()
```

```python
lr.fit(X_train, y_train)
```

```python
lr.predict(X_test)
```

```python
y_test
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score = accuracy_score(y_test, lr.predict(X_test))
```

```python
print(accuracy_score)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(y_test, lr.predict(X_test))
```
