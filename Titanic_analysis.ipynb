```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
df.groupby(["Pclass","Embarked"])["Survived"].mean().plot(kind="bar")
plt.show()
```

```python
# prompt:  diffrents simple charts analysis

# Histogram of Age
plt.figure(figsize=(8, 6))
plt.hist(df['Age'].dropna(), bins=20, color='skyblue', edgecolor='black')
plt.title('Distribution of Passenger Ages')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()

# Bar chart of survival by sex
sns.countplot(x='Sex', hue='Survived', data=df)
plt.title('Survival by Sex')
plt.show()

# Box plot of Fare by Pclass
plt.figure(figsize=(8, 6))
sns.boxplot(x='Pclass', y='Fare', data=df)
plt.title('Fare Distribution by Passenger Class')
plt.show()

# Scatter plot of Age vs. Fare
plt.figure(figsize=(8, 6))
plt.scatter(df['Age'], df['Fare'], alpha=0.5)
plt.title('Age vs. Fare')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.show()

# Interactive scatter plot with Plotly Express
fig = px.scatter(df, x='Age', y='Fare', color='Survived',
                 title='Age vs. Fare with Survival')
fig.show()
```

```python
# prompt: plotly using groupby in titanic

import plotly.express as px

# Group by Pclass and Embarked, calculate average survival rate
grouped_df = df.groupby(["Pclass", "Embarked"])["Survived"].mean().reset_index()

# Create an interactive bar chart
fig = px.bar(
    grouped_df,
    x="Pclass",
    y="Survived",
    color="Embarked",
    barmode="group",
    title="Average Survival Rate by Pclass and Embarked",
)
fig.show()
```

```python
# prompt: scatter plot group by

import matplotlib.pyplot as plt

# Scatter plot of Age vs. Fare, grouped by Pclass
plt.figure(figsize=(8, 6))
for pclass in df['Pclass'].unique():
  subset = df[df['Pclass'] == pclass]
  plt.scatter(subset['Age'], subset['Fare'], label=f'Pclass {pclass}', alpha=0.7)

plt.title('Age vs. Fare, Grouped by Pclass')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.legend()
plt.show()
```
